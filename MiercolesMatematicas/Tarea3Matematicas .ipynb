```python
print("2x âˆ’ y + 2z = 3\n2x + 3y âˆ’ z = 0\nx + 2y + 2z = âˆ’12")
y = np.linspace(-3, 6)
x2, y2 = np.meshgrid(x, y)
z1 = (3-2*x2+y2)/2
z2 = (-2*x2-3*y2)/-1
z3 = (-12-x2-2*y2)/2
fig = plt.figure()
ax = fig.add_subplot(projection="3d") 
ax.plot_surface(x2, y2, z1)
ax.plot_surface(x2, y2, z2)
ax.plot_surface(x2, y2, z3)
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_zlabel("z")
ax.set_title("SISTEMA DE ECUACIONES 3")
```

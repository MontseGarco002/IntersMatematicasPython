```python
#M√âTODO NEWTON
def newton_iters(xn,num_iteraciones):
    iteraciones = [xn] #crea una lista de x0
    errores = [abs(xn-1.)] #Valores de error, se va aproximando a 0
    for a in range(num_iteraciones):
        xn= xn-(xn*xn-1)/(2*xn)
        iteraciones.append(xn)
        errores.append(abs(xn-1))
    return iteraciones,errores
```

```python
it,er = newton_iters(2,5)
fig,axs = plt.subplots(1,2,tight_layout = True) #fila, columna
#tight_layout = True acomoda dos graficas en la misma pantalla
axs[0].plot(it,"x")
axs[0].set_title("Iteraciones")
axs[0].set_xlabel("i")
axs[0].set_ylabel("xi")

axs[1].semilogy(er,"o") #modificacion de rango semilog
axs[1].set_title("Error")
axs[1].set_xlabel("i")
axs[1].set_ylabel("xi")
```

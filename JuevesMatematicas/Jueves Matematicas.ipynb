```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Semilla: base de algoritmos 

#Con el numero fijo, dara lo mismo 
rnd = np.random.default_rng(45671) #Semilla
datos = np.arange(15) #Rango de numeros
print(datos)


#Probabilidad de que salga un numero
probabilidades = np.array(
    [0.3, 0.2, 0.1, 0.05, .05, 0.05, 0.05, 0.025,
    0.025, 0.025, 0.025, 0.025, 0.025, 0.025, 0.025]
)

#Verifica si la probabilidad que tu tienes es correcta
assert round(sum(probabilidades), 10) == 1.0, "la probabilidad debe sumar 1"

seleccion = rnd.choice(datos,p=probabilidades, replace=True)
print(seleccion)

seleccion_datos= rnd.choice(datos,p=probabilidades, replace=True,size = (4,3))
print(seleccion_datos)
#size= columnas, filas

```

```python
rndDos = np.random.default_rng(2456) #Semilla

rndFlotantes = rndDos.random(size=(3,3)) #.random numeros flotantes entre 0 y 3
print(rndFlotantes,"\n")

rndF=np.random.uniform(1.2,1.60,size=10) #con rango de 1.2,1.60
#uniform = la misma probabilidad a todos
print(rndF,"\n")

rndenteros = rndDos.integers(5,15,size=8)
print(rndenteros,"\n")

distribucion = rndDos.integers(1,10,size=20)

fig,ax=plt.subplots()
ax.hist(distribucion,linewidth=0.5, edgecolor="white")
ax.set_title("Histograma de números")
ax.set_xlabel("Valor")
ax.set_ylabel("Frecuencia")
ax.grid()
plt.show()
print(distribucion)


```

```python
rnd = np.random.default_rng(12345)
mu=5.0 #valor medio 
sigma = 3 #desviación estandar

rnds = rnd.normal(loc=mu,scale=sigma,size=1000)
fig,ax=plt.subplots()
ax.hist(rnds,bins=20)
plt.show()
```
